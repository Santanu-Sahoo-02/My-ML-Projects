## Data CLeaning

```python
df[df.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Remove the null values, also rest the index
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]] # this is also repeated
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in column names
df.columns=df.columns.str.strip()
```

```python
df.columns
```

## Change the requires columns as int data type

```python
df[['month','day','year','Temperature', 'RH', 'Ws']] = df[['month','day','year','Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

## Changing the other columns to float data type

```python
objects=[features for features in df.columns if df[features].dtypes=='O'] #list comprehension
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
## Encode the categories in classes
df['Classes']=np.where(df['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
## Monthly Fire Analysis
dftemp=df.loc[df['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp)

plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title("Fire Analysis of Sidi- Bel Regions",weight='bold')
```

```python
## Monthly Fire Analysis
dftemp1=df.loc[df['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp1)

plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title("Fire Analysis of Brjaia Regions",weight='bold')
```

### It's observed that August and September had the most number of forest fires for Sidi-Bel Region. 
### Most of the fires happened in August and very high Fires happened in only 3 months - June, July and August.
### Less Fires was on September in Brjaia region.
