```python
# iris flower classification using sVM
from sklearn.datasets import load_iris
iris= load_iris()
```

```python
# display info about the dataset
dir(iris) # display the attribute names of the dataset
```

```python
iris.data # display data
```

```python
iris.target #0 - satosa, 1 - versicolor, 2 - virginica
```

```python
iris.target_names # flower names
```

```python
iris.feature_names # feature column names
```

```python
# convert iris data set into a dataframe
import pandas as pd
df=pd.DataFrame(iris.data, columns= iris.feature_names)
df.head ()
```

```python
# add target column also to df
df['target'] = iris.target
df.head()
```

```python
# to display pair plots
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# pair plots. hue=' target' -> change color based on target flower
# this uses blue, red and green palette
sns.pairplot(df, hue='target' , palette="brg")
plt.show()
```

```python
#from the above graphs, we can understand that
# We can separate the datapoints using a line.
```

```python
#x represents all col umns except the target column
x = df.drop(['target'], axis ='columns')
# y represents the target column
y = df.target
```

```python
# let us split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
# train the svm model
from sklearn.svm import SVC 
model = SVC() # C=1.0, kernel='rbf', gamma='scale'
model.fit(x_train, y_train)
```

```python
# find accuracy level
model.score (x_test, y_test) 
```

```python
# take a random sample for prediction
# in the data, 0 to 49 =setosa,
# 50 to 99 = versicolor, 1O0 to l49 = virginica
model.predict([iris.data[50]]) # array([1]) -> versicolor
```

```python
# let us give our values for sepal 1,w and petal 1,w
model.predict ([[7.7, 2.6, 6.9, 2.3]]) # array([2]) -> virginica
```
