```python
# classification of heart disease with various models
# create the data frame from the dataset
import pandas as pd
```

```python
#take column names in a 1ist
colnames = ['age','sex', 'cp' ,'trestbps', 'chol', 'fbs', 'restecg' 'thalach','exang' , 'oldpeak', 'slope', 'ca', 'thal', 'num']
```

```python
# create data trame using the data file with column names
df = pd.read_csv("processed.cleveland.data", sep=',', names=colnames)
df
```

```python
# convert num values into 0 or 1 only. call map(O
df['num'] = df.num.map ({0: 0, 1: 1, 2: 1, 3: 1, 4: 1})
df
```

```python
# co1lect input data into x and target column into y
x= df.iloc[:, :-1].values # all cols except 1ast one
y = df.iloc[:, -1].values # take only the last col
```

```python
# divide the data into train and test data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
# Support Vector Classifier Model
# Convert data using standard scaler for SVC model
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train1 = sc.fit_transform(x_train)
x_test1 = sc.transform(x_test)
```

```python
from sklearn.svm import SVC
model=SVC(kernel='rbf')
model.fit(x_train1,y_train)
```

```python
testscore = model.score (x_test1, y_test)
trainscore = model.score (x_train1, y_train)
print ("Score on train data= %.2f" % trainscore)
print("Score on test data= %.2f" % testscore)
```

```python
#Gaussian Naive Bayes Model
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
testscore = model.score(x_test, y_test)
trainscore = model.score (x_train, y_train)
print("score on train data= %.2f" % trainscore)
print("score on test data= %.2f" % testscore)
```

```python
# Logistic Regression Model1
from sklearn.linear_model import LogisticRegression
model = LogisticRegression (solver='liblinear')
model.fit(x_train, y_train)
testscore = model.score (x_test, y_test)
trainscore = model.score(x_train, y_train)
print("Score on train data= %.2f" % trainscore)
print("Score on test data= %.2f" % testscore)
```

```python
# Decision Tree Model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
testscore = model.score(x_test, y_test)
trainscore = model.score (x_train, y_train)
print ("score on train data= %.2f" % trainscore)
print("Score on test data= %.2f" % testscore)
```

```python
# Random Forest Model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train, y_train)
testscore = model.score (x_test, y_test)
trainscore = model.score (x_train, y_train)
print("Score on train data= %.2f" % trainscore)
print("Score on test data= %.2f" % testscore)
```
